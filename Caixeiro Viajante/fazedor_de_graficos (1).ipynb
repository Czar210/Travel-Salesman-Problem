# 20 municipios

```python
import matplotlib.pyplot as plt


metodos = [
    'Nearest Neighbour',
    'Random Walk Best',
    'Random Walk Worst',
    'Random 5 Neighbours Best',
    'Random 5 Neighbours Worst',
    'Random 3 Neighbours Best',
    'Random 3 Neighbours Worst',
    'Lin e Kerninghan',
    'Record to Record',
    'Simulated Annealing'
]

valores = [1941, 2431, 5895, 2017, 4471, 1814, 4645, 1829, 1714, 1714]


metodos_valores = sorted(zip(metodos, valores), key=lambda x: x[1])
metodos_ordenados, valores_ordenados = zip(*metodos_valores)


plt.figure(figsize=(12, 6))
bars = plt.bar(metodos_ordenados, valores_ordenados, color=['cyan', 'blue', 'purple', 'green', 'orange', 'magenta', 'red', 'yellow', 'brown', 'grey'])

for bar, valor in zip(bars, valores_ordenados):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{valor}', va='bottom')

plt.xlabel('Mtodos')
plt.ylabel('Valores')
plt.title('Comparao de Mtodos (Ordenado do Menor para o Maior)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


plt.show()
```
